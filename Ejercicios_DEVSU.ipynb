```python
# Ejercicio # 1
conn = sqlite3.connect('test.db')


####PROGRAME SU RESPUESTA DENTRO DEL CONN.EXECUTE ENTRE LAS COMILLAS

# Query Tabla de Posiciones Qatar 2022
# Parte 1 Separa los nombres de los dos equipos
# Se calcula los según ganó, o empató (derrota no se considera por que sumaría 0) y los goles de diferencia
# Devolviendo tabla de posiciones final agrupado y ordenado

cursor = conn.execute('''
WITH desglose AS (
  -- Equipo local
  SELECT
    REPLACE(SUBSTR (partido, 1, INSTR(partido, ' vs.') - 1), '.', '') AS equipo, grupo,
    CAST(SUBSTR(resultado, 1, INSTR(resultado, ',') - 1) AS INTEGER) AS gf,
    CAST(SUBSTR(resultado, INSTR(resultado, ',') + 1) AS INTEGER) AS gc
    FROM Partidos

  UNION ALL
  -- Equipo visitante

  SELECT
    REPLACE(SUBSTR(partido, INSTR(partido, 'vs. ') + 4), '.', '') AS equipo, grupo,
    CAST(SUBSTR(resultado, INSTR(resultado, ',') + 1) AS INTEGER) AS gf,
    CAST(SUBSTR(resultado, 1, INSTR(resultado, ',') - 1) AS INTEGER) AS gc
    FROM Partidos
  )

  -- Agrupamos y calculamos los punto
  SELECT
    UPPER(TRIM(equipo)) AS equipo, UPPER(TRIM(grupo)) AS grupo,
    SUM(CASE
      WHEN gf > gc THEN 3 WHEN gf = gc THEN 1 ELSE 0
    END) AS total_puntos, SUM(gf - gc) AS gol_diferencia
    FROM desglose
    GROUP BY UPPER(TRIM(equipo)), UPPER(TRIM(grupo))
    ORDER BY grupo ASC, total_puntos DESC, gol_diferencia DESC;

''')

for row in cursor:
  print(row)
conn.close()
```

```python
# Ejercicio # 2
conn = sqlite3.connect('test.db')

####PROGRAME SU RESPUESTA DENTRO DEL CONN.EXECUTE ENTRE LAS COMILLAS

##Cruces para los siguientes partidos
##Se arma el string para que coincida con la tabla y cruzar los datos
## con los equipos de los primeros y sgundos de cada grupo

cursor = conn.execute('''
WITH posiciones AS (
  SELECT equipo, grupo,
  ROW_NUMBER()
  OVER( PARTITION BY grupo ORDER BY puntos DESC, gol_diferencia DESC) as puesto
  FROM Resultados_Qatar
)

  SELECT p1.equipo || ' vs. ' || p2.equipo AS partido, cl.fecha, cl.hora, cl.sede
    FROM posiciones p1, posiciones p2
    JOIN Clasificados cl
      ON lower(cl.partido) = '1º del ' || lower(p1.grupo) || ' vs. 2º del ' || lower(p2.grupo)
    WHERE p1.puesto = 1 AND p2.puesto = 2;
''')

for row in cursor:
  print(row)
conn.close()
```

```python
# Ejercicio # 3
conn = sqlite3.connect('test.db')

####PROGRAME SU RESPUESTA DENTRO DEL CONN.EXECUTE ENTRE LAS COMILLAS
##Obtiene el quintil (20% superior) de los equipos y presenta
##Se calcula total de equipos y lo divide para 5

cursor = conn.execute('''
SELECT equipo, grupo, puntos
  FROM Resultados_Qatar
  ORDER BY puntos DESC
  LIMIT (SELECT COUNT(*) / 5 FROM Resultados_Qatar)
                      ;''')
for row in cursor:
  print(row)
conn.close()
```

```python
from typing import cast
import sys
from pyspark import SparkConf, SparkContext
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import explode, col, desc, avg, min, max, rank, asc, round
from pyspark.sql import Window
from math import sqrt
from pyspark.sql.functions import udf

spark = SparkSession.builder\
        .master("local")\
        .appName("Colab")\
        .config('spark.ui.port', '4050')\
        .getOrCreate()

####### PROGRAME SU RESPUESTA AQUI ##############
from pyspark.sql.functions import asc_nulls_last

####### PROGRAME SU RESPUESTA AQUI ##############

#Respuesta 5.1
#===============================================
#Cargamos el JSON de Centros de Madrid, convertimos las coordenadas a números decimales y quitamos los que no tengan datos
df_madrid = spark.read.option("multiline", "true").json("/content/CENTROS_EDUCATIVOS_MADRID.json") \
                 .withColumn("cx", col("direccion_coor_x").cast("double")) \
                 .withColumn("cy", col("direccion_coor_y").cast("double")) \
                 .filter(col("cx").isNotNull() & col("cy").isNotNull())

#Respuesta 5.2.1
#===============================================
#Calculamos el promedio de X y Y por cada tipo de titularidad y lo unimos a la tabla original
df_proms = df_madrid.groupBy("centro_titularidad").agg(avg("cx").alias("px"), avg("cy").alias("py"))
df_unido = df_madrid.join(df_proms, "centro_titularidad", "left")

#Respuesta 5.2.2
#===============================================
#Creamos la función UDF para calcular la distancia de cada colegio hacia el punto promedio del grupo
def calc_distancia(x1, y1, x2, y2):
    try:
        return ((float(x1) - float(x2))**2 + (float(y1) - float(y2))**2) ** 0.5
    except:
        return None

udf_dist = udf(calc_distancia, DoubleType())
df_con_dist = df_unido.withColumn("distancia", udf_dist(col("cx"), col("cy"), col("px"), col("py")))

#Respuesta 5.2.3
#===============================================
#Usamos una ventana para encontrar el colegio con la distancia más pequeña en cada grupo y guardamos la vista
ventana = Window.partitionBy("centro_titularidad").orderBy("distancia")

df_final = df_con_dist.withColumn("fila", row_number().over(ventana)) \
                      .filter(col("fila") == 1) \
                      .drop("fila", "px", "py", "cx", "cy")

df_final.select("centro_titularidad", "centro_nombre", "distancia").show(truncate=False)
df_final.createOrReplaceTempView("vista_centricos")

#################################################
```
